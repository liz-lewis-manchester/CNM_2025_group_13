```python
# Choosing a small system for testing
N_x = 5

A = np.array([2.0, 2.0, 2.0, 2.0])
B = np.array([-1.0, -1.0, -1.0, -1.0])

C_next = np.zeros(N_x)
C_next[0] = 1.0  # boundary condition

F = np.array([1.0, 1.0, 1.0, 1.0])
```

```python
#Running the test
C_next=forward_substitution(A, B, F, C_next)

print(C_next)
```

```python
# Manual expected solution
expected = np.zeros(N_x)
expected[0] = 1.0
expected[1] = (1 - (-1)*1) / 2
expected[2] = (1 - (-1)*expected[1]) / 2
expected[3] = (1 - (-1)*expected[2]) / 2
expected[4] = (1 - (-1)*expected[3]) / 2

assert np.allclose(C_next, expected)

print("Task 3 forward substitution test passed.")
```
